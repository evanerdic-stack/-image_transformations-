```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membaca, mentransformasi, dan menampilkan gambar
def image_transformations(image_path):
    # Baca gambar dalam skala abu-abu
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Periksa apakah gambar berhasil dimuat
    if img is None:
        raise ValueError("Gambar tidak ditemukan atau path tidak benar.")

    # Transformasi negatif
    img_neg = 255 - img

    # Transformasi logaritmik
    max_pixel_val = float(np.max(img)) # Ensure max_pixel_val is float to prevent overflow in 1 + max_pixel_val
    if max_pixel_val == 0:
        # Jika gambar benar-benar hitam, transformasi logaritmik tidak akan mengubahnya.
        # Set hasilnya menjadi gambar hitam juga.
        img_log = np.zeros_like(img, dtype=np.uint8)
    else:
        c = 255 / np.log(1 + max_pixel_val)
        # Pastikan img diubah ke float sebelum operasi logaritmik untuk menghindari masalah tipe data
        img_log = c * np.log(1 + img.astype(float))
        # Klip nilai ke rentang 0-255 sebelum konversi ke uint8 untuk mencegah 'invalid value encountered in cast'
        img_log = np.uint8(np.clip(img_log, 0, 255))

    # Plot gambar dan histogram
    plt.figure(figsize=(16, 8))

    plt.subplot(2, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Gambar Asli')
    plt.subplot(2, 3, 4)
    plt.hist(img.flatten(), bins=256, range=(0, 256), color='r') # Menggunakan named arguments
    plt.title('Histogram Asli')

    plt.subplot(2, 3, 2)
    plt.imshow(img_neg, cmap='gray')
    plt.title('Gambar Negatif')
    plt.subplot(2, 3, 5)
    plt.hist(img_neg.flatten(), bins=256, range=(0, 256), color='g') # Menggunakan named arguments
    plt.title('Histogram Negatif')

    plt.subplot(2, 3, 3)
    plt.imshow(img_log, cmap='gray')
    plt.title('Gambar Logaritmik')
    plt.subplot(2, 3, 6)
    plt.hist(img_log.flatten(), bins=256, range=(0, 256), color='b') # Menggunakan named arguments
    plt.title('Histogram Logaritmik')

    plt.show()

# Contoh penggunaan:
image_file_path = 'your_image.jpg.jpeg'  # Ganti dengan path gambar Anda yang benar
image_transformations(image_file_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Fungsi untuk menerapkan filter yang berbeda dan menampilkan hasilnya
def apply_convolution_filters(image_path):
    # Baca gambar dalam skala abu-abu
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Periksa apakah gambar berhasil dimuat
    if img is None:
        raise ValueError("Gambar tidak ditemukan atau path tidak benar.")

    # Definisikan kernel
    lpf_kernel = np.array([[1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9],
                           [1/9, 1/9, 1/9]])
    hpf_kernel = np.array([[-1, -1, -1],
                           [-1, 8, -1],
                           [-1, -1, -1]])
    bpf_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

    # Terapkan filter menggunakan konvolusi
    img_lpf = convolve2d(img, lpf_kernel, mode='same', boundary='fill', fillvalue=0).astype(np.uint8)
    img_hpf = convolve2d(img, hpf_kernel, mode='same', boundary='fill', fillvalue=0).astype(np.uint8)
    img_bpf = convolve2d(img, bpf_kernel, mode='same', boundary='fill', fillvalue=0).astype(np.uint8)

    # Hitung nilai piksel rata-rata
    avg_original = np.mean(img)
    avg_lpf = np.mean(img_lpf)
    avg_hpf = np.mean(img_hpf)
    avg_bpf = np.mean(img_bpf)

    # Tampilkan gambar
    plt.figure(figsize=(16, 4))

    plt.subplot(1, 4, 1)
    plt.imshow(img, cmap='gray')
    plt.title(f'Asli (Rata-rata: {avg_original:.2f})')

    plt.subplot(1, 4, 2)
    plt.imshow(img_lpf, cmap='gray')
    plt.title(f'LPF (Rata-rata: {avg_lpf:.2f})')

    plt.subplot(1, 4, 3)
    plt.imshow(img_hpf, cmap='gray')
    plt.title(f'HPF (Rata-rata: {avg_hpf:.2f})')

    plt.subplot(1, 4, 4)
    plt.imshow(img_bpf, cmap='gray')
    plt.title(f'BPF (Rata-rata: {avg_bpf:.2f})')

    plt.show()

# Contoh penggunaan:
apply_convolution_filters('your_image.jpg.jpeg')  # Ganti 'your_image.jpg' dengan path gambar Anda
```
